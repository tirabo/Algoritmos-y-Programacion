## Otras funciones que podés usar:

*   `bgcolor(color)`
*   `color(color)`
*   `forward(units)`
*   `backward(units)`
*   `speed(speed)`, `speed` debe ser un número entre `1` y `13`
*   `right(degrees)`
*   `left(degrees)`
*   `face(degrees)`, para apuntar en cierta dirección
*   `penup()`, levanta el lápiz, para poder desplazarse sin dejar huella
*   `pendown()`, apoya el lápiz, para volver a dibujar después de haberlo levantado
*   `width(width)`, para cambiar el grosor del lápiz
*   `hideturtle()`, para que no se vea la tortuga
*   `showturtle()`, para que se la vuelva a ver

Más funciones se presentan al final de [ColabTurtle](https://github.com/tolgaatam/ColabTurtle).

Captura de pantalla de 2021-03-21 08-00-15.png

Acá aparece un bloque repetir. La posibilidad de repetir, iterar instrucciones, no es una característica exclusiva del artista. Por ello, no está en la lista de "otras funciones que podés usar" que se mostró recién.

Todos los lenguajes de programación imperativa tienen alguna manera de repetir. Cualquiera sea el número natural ```n``` La manera más sencilla de traducir el bloque 
```
repetir n veces ...
```
a Python, es
```
for _ in range(n):
    ...
```


````
# configuración pizarra inicial
initializeTurtle()
bgcolor('white')
color('black')
right(90)

# solución del ejercicio 3 del artista
for _ in range(4):
    forward(100)
    right(90)
hideturtle()
````

Los dos puntos (```:```) al final de la línea 8, y la sangría en las líneas 9 y 10 son fundamentales. La sangría determina cuáles son las líneas cuyas instrucciones deben iterarse. Se recomienda usar 4 espacios como sangría. En Colab puede configurarse tocando la rueda dentada (arriba derecha) y  luego "Editor".

Hemos repetido tantas veces las 4 líneas de código que generan una pizarra vacía inicial. Ya es hora de **definir una función** que agrupe esas 4 líneas y nos ahorre el trabajo de repetirlas (y tener que reconocerlas) cada vez.

````
def pizarra_vacia():
    initializeTurtle()
    bgcolor('white')
    color('black')
    right(90)
````

Ahora `pizarra_vacia()` es una función más,  como lo es `initializeTurtle()`, solo que la hemos definido nosotros. Veremos en la próxima clase como se definen funciones en Python con cierto detalle.

Esta definición también puede desdoblarse así:

````
# configuración pizarra inicial
pizarra_vacia()

# solución del ejercicio 2 del artista 2
for _ in range(4):
    col = choice(['yellow', 'orange', 'red', 'green', 'blue', 'purple', 'grey', 'black']) # guarda el color elegido en la variable col
    color(col)                                                                            # adopta el color determinado por col
    for _ in range(3): # dibujar triángulo
        forward(100)
        right(120)
    right(90)
````
